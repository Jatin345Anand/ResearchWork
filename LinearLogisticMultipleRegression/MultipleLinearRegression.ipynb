```python
def COST(X,Y,b):
    m = len(X)
    Cost = np.sum((X.dot(b)-Y)**2)/(2*m)
    return Cost
```

```python
def GD(X,Y,b,EPOCH,ALPHA):
    Cost_History = [0]*EPOCH
    m = len(read)
    for i in range(EPOCH):
        h = X.dot(b)
        loss =h - Y
        Gredient = X.T.dot(loss)/m
        b = b - ALPHA*Gredient
        Cost_History[i] = COST(X,Y,b)
    return Cost_History,b
```

```python
COSTHISTORY,b = GD(X,y,b,nb_epoch,alpha)
```
